```python
def transfor(x):
    if x.find("M")>=0:
        x=float(x.replace("M",""))*1000000
    elif x.find("k")>=0:
        x=float(x.replace("k",""))*1000
    elif x=="Varies with device":
        x="x"
    return x
```

```python
transfor(df["Size"].iloc[0])
```

```python
if 'adrian'.find("x")>=0:
    print("verdadero")
```

```python
df["Size"]=df["Size"].apply(transfor)
```
