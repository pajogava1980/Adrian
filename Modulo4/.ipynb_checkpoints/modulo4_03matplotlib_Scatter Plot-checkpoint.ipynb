<h1 align="center">Scatter Plot</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('FB.csv')
df.head()
```

<h2>Grafico Scatter con el metodo plot</h2>

```python
plt.plot(df.index,df.Close, "o")
plt.grid()
```

```python
plt.plot(df.index,df.Close, "+")
plt.grid()
```

```python
plt.plot(df.index,df.Close, "v")
plt.grid()
```

```python
plt.plot(df.index,df.Close, "rv")
plt.grid()
```

```python
plt.plot(df.index,df.Close, "bP", markersize=10)
plt.grid()
```

<h2>Grafico Scatter con el metodo scatter</h2>

```python
plt.scatter(df.index,df.Close)
plt.grid()
```

```python
plt.scatter(df.index,df.Close,s=75)
plt.grid()
```

```python
plt.scatter(df.index,df.Close,s=75, c="m")
plt.grid()
```

```python
plt.scatter(df.index,df.Close,s=75, c="m", marker="v")
plt.grid()
```

```python
plt.scatter(df.index,df.Close,s=75, c="m",edgecolor="black",linewidth=2)
plt.grid()
```

```python
plt.scatter(df.index,df.Close,
            s=100, c="m", 
            edgecolor="black", 
            linewidth=1, 
            alpha=0.5)
plt.grid()
```

```python
color=np.random.randint(0,10,251)
```

```python
color
```

```python
plt.scatter(df.index,df.Close, 
            s=100, 
            c=color, 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
plt.grid()
```

```python
plt.scatter(df.index,df.Close, 
            s=100, 
            c=color, 
            cmap="RdPu", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
plt.grid()
```

```python
plt.scatter(df.index,df.Close, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
plt.grid()
```

```python
plt.scatter(df.index,df.Close, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
plt.grid()
```

```python
lista=df.Close.to_list()
```

```python
len(lista)
```

```python
plt.scatter(df.index,df.Close, 
            s=100, 
            c=lista, 
            cmap="PuRd", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
plt.grid()
```

```python
plt.scatter(df.index,df.Close, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
cbar.set_label("Levels")
plt.grid()
```

```python
sizes=np.random.randint(50,200,251)
```

```python
plt.scatter(df.index,df.Close, 
            s=sizes,  
            c=df.Close.to_list(), 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)

plt.grid()
```

```python
plt.scatter(df.index,df.Close, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("Indice")
plt.ylabel("Precio de cierre")

plt.grid()
```

```python
plt.scatter(df.index,df.Close, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("Indice")
plt.ylabel("Precio de cierre")
plt.title("Precio accion Facebook")
```

```python
plt.figure(figsize=(10,10))
plt.grid()
plt.scatter(df.index,df.Close, 
            s=sizes,  
            c=df.index.to_list(), 
            cmap="Greys",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
```
