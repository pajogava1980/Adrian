## <p style="background-color:DodgerBlue;font-family:newtimeroman;font-size:150%;color:white;text-align:center;border-radius:20px 60px;">Figura y Subplot</p>

```python
%matplotlib notebook 
```

```python
fig=plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
data=np.random.randn(50,).cumsum() 
ax1.plot(data, 'k--')
ax2.plot(data, "r-")
ax3.plot(data, "b.")
```

## <p style="background-color:DodgerBlue;font-family:newtimeroman;font-size:150%;color:white;text-align:center;border-radius:20px 60px;">Color y Estilos</p>

```python
x=[1,2,3,4,5]
y=[10,11,12,13,14]

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

#dos formas de dar color y estilo a la línea

#ax.plot(x, y, 'r--')
ax.plot(x, y, linestyle='--', color='r')
```

## <p style="background-color:DodgerBlue;font-family:newtimeroman;font-size:150%;color:white;text-align:center;border-radius:20px 60px;">Etiquetas y Leyendas</p>

```python
%matplotlib notebook
```

```python
fig=plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(data)
location=ax.set_xticks([0,10,20,30,40,50])
label=ax.set_xticklabels(["uno","dos","tres","cuatro","cinco","seis"],rotation=30,fontsize="small")
ax.set_title('Primer grafico')
ax.set_xlabel('Etapas')
ax.set_ylabel('Alguna cosa')
#feature={"title":"Primer grafico otra forma", 
#         "xlabel":"Pasos", 
#         "ylabel":"Numeros"}
#ax.set(**feature)
```

## <p style="background-color:DodgerBlue;font-family:newtimeroman;font-size:150%;color:white;text-align:center;border-radius:20px 60px;">Adicionar Leyendas</p>

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
data1=np.random.randn(1000).cumsum()
data2=np.random.randn(1000).cumsum()
data3=np.random.randn(1000).cumsum()

ax.plot(data1, 'k', label='one')
ax.plot(data2, 'b--', label='two')
ax.plot(data3, 'r.', label='three')
ax.legend()
```

## <p style="background-color:DodgerBlue;font-family:newtimeroman;font-size:150%;color:white;text-align:center;border-radius:20px 60px;">Anotaciones</p>

```python
df=pd.read_csv('FB.csv')
df.head()
```

```python
df=pd.read_csv('FB.csv',index_col=0,parse_dates=True)
#parse_dates: si es True intenta transformar el indice en un tipo fecha o date
```

```python
df.head()
```

```python
c_price=df["Close"]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
c_price.plot(ax=ax, style='k-')
ax.set_title('Facebook stock closing prices')
```

```python
date="2018-12-24"
text="Minimum"
```
