## <p style="background-color:DodgerBlue;font-family:newtimeroman;font-size:150%;color:white;text-align:center;border-radius:20px 60px;">Adicionar Leyendas</p>

```python
data1=np.random.randn(1000).cumsum()
data2=np.random.randn(1000).cumsum()
data3=np.random.randn(1000).cumsum()
```

```python
fig=plt.figure(figsize=(10, 10))
ax=fig.add_subplot(2,1,1)

ax.plot(data1, 'k', label = "primera serie")
ax.plot(data2, 'g--', label = "segunda serie")
ax.plot(data3, 'b-.', label = "tercera serie")

ax.legend()
ax.grid()

ax1=fig.add_subplot(2,1,2)
ax1.plot(data2[800:850], 'g--', label = "segunda serie")
ax1.legend()
ax1.grid()
```
