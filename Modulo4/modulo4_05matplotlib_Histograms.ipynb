<h2 >Como plotear un histograma</h2>

```python
data=np.random.randn(1000)
```

```python
data
```

```python
plt.hist(data)
```

```python
plt.hist(data)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
```

```python
plt.hist(data, edgecolor="black")
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
```

```python
plt.hist(data, edgecolor="black", bins=10)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
```

```python
bins=[-3,-2,-1,0,1,2,3]
```

```python
plt.hist(data, edgecolor="black", bins=bins)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
```

```python
median=np.median(data)
```

```python
plt.hist(data, edgecolor="black", bins=bins)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.axvline(median, color="red")
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-3, 2, 1000)
x3 = np.random.normal(2, 1, 1000)
```

```python
plt.hist(x1, histtype="stepfilled", 
         alpha=0.5, bins=40)
plt.hist(x2, histtype="stepfilled", 
         alpha=0.5, bins=40)
plt.hist(x3, histtype="stepfilled", 
         alpha=0.5, bins=40)
```

<h2>Histograma de dos dimensiones</h2>

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, 
                                     cov, 
                                     10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='YlGnBu')
```

```python
plt.hist2d(x, y, bins=30, cmap='Reds')
cb = plt.colorbar()
cb.set_label('Number of rectangles')
```

<h2>Aplicación con datos reales</h2>

```python
movies=pd.read_csv("imdbratings.txt")
```

```python
movies.head()
```

```python
plt.hist(movies.duration)
```

```python
plt.hist(
    movies.duration, edgecolor="black", color="m")
```

```python
bins=np.arange(70,200,10)
```

```python
plt.hist(movies.duration, edgecolor="black", 
         color="m", bins=bins)
plt.xticks(bins)
```

```python
plt.hist(movies.duration, edgecolor="black", 
         color="m", bins=bins)
plt.xticks(bins)
plt.title("Histograms of movie durations in imdb")
plt.xlabel("Durations")
plt.ylabel("Frequencies")
```
