### Valores propios

```python
matrix = np.array([[1,2],[0,1]])
```

```python
np.linalg.eigvals(matrix)
```

La eigendecomposición consiste simplemente en descomponer los objetos matemáticos en sus partes constitutivas.  Del mismo modo, podemos descomponer las matrices de manera que revelen información sobre sus propiedades funcionales que no es inmediatamente obvia. De esta forma, tomamos una matriz y la descomponemos en vectores propios y valores propios. Estas descomposiciones matriciales son una herramienta útil para reducir una matriz a sus partes constituyentes con el fin de simplificar una serie de operaciones más complejas. De igual forma, desempeña un papel en los métodos utilizados en Machine Learning, como en el método de análisis de componentes principales o PCA.
